Pasé los libros con "boxed set, colection o encyclopedia" a una lista a parte la cual ahora dividiré en 4 categorias Libros individuales estándar, Libros individuales largos, Colecciones y Enciclopedias.

```python
import pandas as pd

# Marcar las colecciones y enciclopedias conocidas
df['is_collection'] = df['Name'].str.contains('Collection|Boxed Set|Complete|Encyclopedia', case=False)

def categorize_book(row):
    if row['is_collection']:
        return 'Collection/Encyclopedia'
    elif row['pagesNumber'] <= 250:
        return 'Short Book'
    elif 251 <= row['pagesNumber'] <= 1000:
        return 'Long Book'
    elif 1001 <= row['pagesNumber'] <= 2000:
        return 'Collection'
    else:
        return 'Encyclopedia'

df['book_category'] = df.apply(categorize_book, axis=1)

# Mostrar las primeras filas para verificar
print(df.head())
```

```python

import pandas as pd

# Cargar el DataFrame con pandas
dataframe = pd.read_csv('C:\\Users\\Fukushima\\Documents\\GitHub\\ProyectoFinalVV\\data\\raw\\booksfiltrados.csv')

def target_encode(dataframe, categorical_col, target_col):
    # Calcular la media del objetivo para cada categoría
    means = dataframe.groupby(categorical_col)[target_col].mean()
    
    # Crear un diccionario de codificación
    encoding = means.to_dict()
    
    # Reemplazar los valores en la columna categórica con la media del objetivo
    dataframe[f'{categorical_col}_encoded'] = dataframe[categorical_col].map(encoding).fillna(0)
    
    return dataframe

# Suponiendo que la columna objetivo se llama 'target'
dataframe = target_encode(dataframe, 'Name', 'Rating')
dataframe = target_encode(dataframe, 'Authors', 'Rating')

```
