*Tengo dos columnas que contienen la misma información y me servirían más en una sola columna, uniré pagesNumber y PagesNumber en una sola columna para tener todo mejor organizado.*

```python
import pandas as pd

# Combinar las dos columnas en una sola
# Si alguna de las dos columnas tiene valores nulos, se llenará con el valor de la otra columna
df['pagesNumber'] = df['pagesNumber'].combine_first(df['PagesNumber'])

# Eliminar la columna duplicada
df = df.drop(columns=['PagesNumber'])

# Verificar el resultado
print(df.head())
```

Al realizar el cuadro anterior pude observar que habían libros que tenían demasiadas páginas iban desde las 100 hasta las más de 4000 entonces decidí filtrar y crear unas columnas adicionales para libros o colecciones con una extensión demasiado grandes, hay incluso enciclopedias con una extención de más de 30000 páginas.


Pasé los libros con "boxed set, colection o encyclopedia" a una lista a parte la cual ahora dividiré en 4 categorias Libros individuales estándar, Libros individuales largos, Colecciones y Enciclopedias.

```python
import pandas as pd

# Marcar las colecciones y enciclopedias conocidas
df['is_collection'] = df['Name'].str.contains('Collection|Boxed Set|Complete|Encyclopedia', case=False)

def categorize_book(row):
    if row['is_collection']:
        return 'Collection/Encyclopedia'
    elif row['pagesNumber'] <= 250:
        return 'Short Book'
    elif 251 <= row['pagesNumber'] <= 1000:
        return 'Long Book'
    elif 1001 <= row['pagesNumber'] <= 2000:
        return 'Collection'
    else:
        return 'Encyclopedia'

df['book_category'] = df.apply(categorize_book, axis=1)

# Mostrar las primeras filas para verificar
print(df.head())
```

```python
df.book_category
```

```python
import pandas as pd

# Filtrar el DataFrame por cada categoría
df_cortos = df[df['book_category'] == 'Short Book']
df_largos = df[df['book_category'] == 'Long Book']
df_colecciones_enciclopedias = df[df['book_category'] == 'Collection/Encyclopedia']

# Análisis descriptivo para libros cortos
print("Análisis de Libros Cortos")
print(df_cortos.describe(include='all'))

# Análisis descriptivo para libros largos
print("\nAnálisis de Libros Largos")
print(df_largos.describe(include='all'))

# Análisis descriptivo para colecciones y enciclopedias
print("\nAnálisis de Colecciones y Enciclopedias")
print(df_colecciones_enciclopedias.describe(include='all'))

# Si quieres ver la distribución de una columna específica, por ejemplo, 'average_rating'
print("\nDistribución de Ratings de Libros Cortos")
print(df_cortos['Rating'].describe())

print("\nDistribución de Ratings de Libros Largos")
print(df_largos['Rating'].describe())

print("\nDistribución de Ratings de Colecciones y Enciclopedias")
print(df_colecciones_enciclopedias['Rating'].describe())
```

```python
# Libros estándar
sns.histplot(data=df[df['book_category'] == 'Short Book'], x="pagesNumber", color='purple')
plt.title('Distribución de número de páginas - Libros cortos')
plt.show()

# Libros largos
sns.histplot(data=df[df['book_category'] == 'Long Book'], x="pagesNumber", color='pink')
plt.title('Distribución de número de páginas - Libros Largos')
plt.show()

# Colecciones y Enciclopedias
sns.histplot(data=df[df['book_category'] == 'Collection/Encyclopedia'], x="pagesNumber", color='green')
plt.title('Distribución de número de páginas - Colecciones y Enciclopedias')
plt.show()
```
